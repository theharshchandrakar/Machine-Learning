```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

set working directory

```python
import os
os.chdir('G://My Drive/#myCDACfiles/MachineLearning/Datasets')
```

load data file into a dataframe

```python
house = pd.read_csv('Housing.csv')
```

Encoding

```python
yes_no_cols = ["driveway", "recroom", "fullbase", "gashw", "airco", "prefarea"]  # adjust to your file
house[yes_no_cols] = house[yes_no_cols].replace({"yes": 1, "no": 0})
```

select features and label columns

```python
X = house.drop(columns='price')
y = house['price']
```

split data into train, test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=35)
```

Model

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

Predict

```python
y_pred = model.predict(X_test)
y_pred
```
