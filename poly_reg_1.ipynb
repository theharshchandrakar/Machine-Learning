### Boston

```python

# Read the Boston dataset and split it into predictors (X) and the target variable (y)
boston = pd.read_csv('Boston.csv')
X = boston.drop('medv', axis=1)
y = boston['medv']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
```

```python
#Degree 2
poly = PolynomialFeatures(degree=2, include_bias=False).set_output(transform='pandas')
X_trn_poly = poly.fit_transform(X_train)   #train set transformed into polynomials
X_tst_poly = poly.transform(X_test)
lr = LinearRegression()
lr.fit(X_trn_poly, y_train)
y_pred = lr.predict(X_tst_poly)
mean_absolute_error(y_test, y_pred)
```

```python
#Degree 3
poly = PolynomialFeatures(degree=3, include_bias=False).set_output(transform='pandas')
X_trn_poly = poly.fit_transform(X_train)   #converted X to x, x^2, x^3
X_tst_poly = poly.transform(X_test)
lr = LinearRegression()
lr.fit(X_trn_poly, y_train)
y_pred = lr.predict(X_tst_poly)
mean_absolute_error(y_test, y_pred)
```

```python
#Degree 4
poly = PolynomialFeatures(degree=4, include_bias=False).set_output(transform='pandas')
X_trn_poly = poly.fit_transform(X_train)   #converted X to x, x^2, x^3
X_tst_poly = poly.transform(X_test)
lr = LinearRegression()
lr.fit(X_trn_poly, y_train)
y_pred = lr.predict(X_tst_poly)
mean_absolute_error(y_test, y_pred)
```

### Housing

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
housing = pd.read_csv('Housing.csv')
```

```python
cat_house = housing[ housing.columns[housing.dtypes == 'object'] ]
num_house = housing[ housing.columns[housing.dtypes != 'object'] ]
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False).set_output(transform='pandas')
ohe_house = ohe.fit_transform(cat_house)
house = pd.concat([num_house, ohe_house],axis=1)
X, y = house.drop('price', axis=1), house['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
house
```

```python
poly = PolynomialFeatures(degree=1, include_bias=False).set_output(transform='pandas')
X_trn_poly = poly.fit_transform(X_train)   #train set transformed into polynomials
X_tst_poly = poly.transform(X_test)
lr = LinearRegression()
lr.fit(X_trn_poly, y_train)
y_pred = lr.predict(X_tst_poly)
mean_absolute_error(y_test, y_pred)
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False).set_output(transform='pandas')
X_trn_poly = poly.fit_transform(X_train)   #train set transformed into polynomials
X_tst_poly = poly.transform(X_test)
lr = LinearRegression()
lr.fit(X_trn_poly, y_train)
y_pred = lr.predict(X_tst_poly)
mean_absolute_error(y_test, y_pred)
```

```python
poly = PolynomialFeatures(degree=3, include_bias=False).set_output(transform='pandas')
X_trn_poly = poly.fit_transform(X_train)   #train set transformed into polynomials
X_tst_poly = poly.transform(X_test)
lr = LinearRegression()
lr.fit(X_trn_poly, y_train)
y_pred = lr.predict(X_tst_poly)
mean_absolute_error(y_test, y_pred)
```

```python
poly = PolynomialFeatures(degree=4, include_bias=False).set_output(transform='pandas')
X_trn_poly = poly.fit_transform(X_train)   #train set transformed into polynomials
X_tst_poly = poly.transform(X_test)
lr = LinearRegression()
lr.fit(X_trn_poly, y_train)
y_pred = lr.predict(X_tst_poly)
mean_absolute_error(y_test, y_pred)
```

### Transformer

```python
from sklearn.compose import ColumnTransformer
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False).set_output(transform='pandas')
col_trnf = ColumnTransformer([('OHE',ohe, list(housing.columns[housing.dtypes==object]))],
                             remainder = 'passthrough',
                             verbose_feature_names_out = False)
col_trnf = col_trnf.set_output(transform='pandas')
col_trnf.fit_transform(housing)
```
